```python

distb = pd.read_sql_query("""Select count(*) as projectcount,year, case 
                           when SOW>0 and SOW<= 40000 then 'A(0-40K)' 
                           when SOW>40000 and SOW<= 80000 then 'B(80-40K)'
                           when SOW>80000 and SOW<= 120000 then 'C(120-80K)'
                           when SOW > 120000 and SOW <= 220000 then 'D(220-120k)'
                           when SOW > 220000 and SOW <= 320000 then 'E(320-220K)'
                           when SOW > 320000 and SOW <=420000 then 'F(420-320k)'
                           when SOW > 420000 and SOW <= 520000 then 'G(520-420k)'
                           else 'H(521k) plus' end as SOWRange
                           from  
                             (Select j.jobcode,datepart(year,j.projStartDate) year, datepart(month,j.projStartDate) month,
                                  ((case WHEN j.FixedContractAmount > 0 then j.FixedContractAmount else 0.0 end)*ISNULL(ccr.conversionFactor,1.0)) AS SOW, 
                                  (ISNULL(j.SOWTotalCalc,0.0)*ISNULL(ccr.conversionFactor,1.0)) as Budget
                              from AllOpenJobs j INNER JOIN
                              RateCardTables as rct on rct.RateCardTableID = j.RateCardTableID LEFT  JOIN
                              FIN_CurrencyConversionRate AS ccr on ccr.currencyToID = 1 and ccr.currencyFromID = rct.currencyID and 
                              ccr.effectiveDate <= j.projStartDate AND ISNULL(ccr.cutoffDate,'1/1/2079') > j.projStartDate
                             where  j.retainerID=0 and j.IsNonBillable=0) job

                            where SOW > 100
                            group by 

                            case when SOW>0 and SOW<= 40000 then 'A(0-40K)' 
                           when SOW>40000 and SOW<= 80000 then 'B(80-40K)'
                           when SOW>80000 and SOW<= 120000 then 'C(120-80K)'
                           when SOW > 120000 and SOW <= 220000 then 'D(220-120k)'
                           when SOW > 220000 and SOW <= 320000 then 'E(320-220K)'
                           when SOW > 320000 and SOW <=420000 then 'F(420-320k)'
                           when SOW > 420000 and SOW <= 520000 then 'G(520-420k)'
                           else 'H(521k) plus' end, 
                            year 
                            order by year desc, SOWRange """,con)
```

```python
import matplotlib.pyplot as plot
plot1= distb[(distb['year'] == 2018)]
plot1.plot(kind='bar',x='SOWRange',y='projectcount', title ="SOW Distribution 2018",figsize=(21, 15), legend=True, fontsize=12)
plot.show() 
```

```python
plot1= distb[(distb['year'] == 2017)]
plot1.plot(kind='bar',x='SOWRange',y='projectcount', title ="SOW Distribution 2017",figsize=(21, 15), legend=True, fontsize=12)
plot.show()
```
