```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2,random_state=0)
```

```python
x_test= df1.apply(pd.to_numeric,errors='coerce')
x_test = x_test.fillna(0)
```

```python
x_test_2d = tsne.fit_transform(x_test)
```

```python
import numpy as np
from sklearn.mixture import GaussianMixture

bic = np.zeros(20)
n = np.arange(1,21)
models = []
#loop through each number of Gaussians and compute the BIC, and save the model
for i,j in enumerate(n):
    #create mixture model with j components
    gmm = GaussianMixture(n_components=j)
    #fit it to the data
    gmm.fit(x_test_2d)
    #compute the BIC for this model
    bic[i] = gmm.bic(x_test_2d)
    #add the best-fit model with j components to the list of models
    models.append(gmm)
```

```python
import matplotlib.pyplot as plt
plt.plot(n,bic)
```

```python
import visuals as vs
from sklearn.metrics import silhouette_score

clusterer = GaussianMixture(n_components=10).fit(x_test_2d)

# Predict the cluster for each data point
preds = clusterer.predict(x_test_2d)

# ind the cluster centers
centers = clusterer.means_

#  Predict the cluster for each transformed sample data point
#sample_preds = clusterer.predict(pca_samples)

# Calculate the mean silhouette coefficient for the number of clusters chosen


score= silhouette_score(x_test_2d,preds)
   
```

```python
x_test_2d_df = pd.DataFrame(x_test_2d,columns=['Dim1','Dim2'])
```

```python
import matplotlib.cm as cm

predictions = pd.DataFrame(preds, columns = ['Cluster'])
plot_data = pd.concat([predictions, x_test_2d_df], axis = 1)

# Generate the cluster plot
fig, ax = plt.subplots(figsize = (14,8))
# Color map
cmap = cm.get_cmap('gist_rainbow')

# Color the points based on assigned cluster
for i, cluster in plot_data.groupby('Cluster'):   
    cluster.plot(ax = ax, kind = 'scatter', x = 'Dim1', y = 'Dim2', \
    color = cmap((i)*1.0/(len(centers)-1)), label = 'Cluster %i'%(i), s=30);
    
for i, c in enumerate(centers):
    ax.scatter(x = c[0], y = c[1], color = 'white', edgecolors = 'black', \
    alpha = 1, linewidth = 2, marker = 'o', s=200);
    ax.scatter(x = c[0], y = c[1], marker='$%d$'%(i), alpha = 1, s=100);
```
